## 3.2 Distribution of Target Variable

```python
fig, ax = plt.subplots(figsize = (10, 8))
values = dataset['fraud'].value_counts().values
explode = [0, 0.2]
labels = ['Non-Fraudulent', 'Fraudulent']

# Seaborn color palette to plot pie chart
colors = sns.color_palette('pastel')

# create pie chart using matplotlib
plt.pie(values, labels=labels, colors=['green', 'red'],
        autopct='%.0f%%', explode=explode, shadow=False, rotatelabels='False', 
        textprops={'fontsize': 14, 'color': 'black'},
        pctdistance = 0.7,
       )
plt.show()
```

The chart indicates that there is a class inbalance in the dataset. About 99% of the entire transactions are non-fraudulent. Only a percent of the total transactions are fraudulent.  The implication of this inbalance on predictive analysis is the creation of a biased model. Models trained on this dataset would have been trained on so much non-fraudulent transactions and very little non-fradulent ones that it would find it very difficult to identify a fraudulent transaction when it is tasked with identifying one. Hence, the dataset should be balanced before usage for modelling.

## 3.5 Analysis of Fraudulent transactions 

Among the main objective of this project is the identification of fraudulent transanctions. This section explores the transactions tagged fraudulent in an attempt to discover common trait and patterns in them.

```python
# filter out the fraudulent transactions
fraudulent_df = dataset[dataset['fraud'] == 1]


# filter out the non fraudulent transactions
non_fraudulent_df = dataset[dataset['fraud'] == 0]
```

```python
# total number of faudulent transactions
fraudulent_df.shape
```

```python
def ask_fraudulent(feature):
    '''
        Returns the fraudulent count, total transaction and their ratio indexed by feature
        
        Parameter
        feature : string - The feature on which the dataset is grouped and counted
        
        return: dataframe - A table containing the count of fraudulent transaction 
                            and their proportion sorted by proportion
    '''
    # get a list of unique entries
    fraud_cols = list(fraudulent_df[feature].unique()) 
    # count fraudulent transanctions by requested feature
    count_frauds = fraudulent_df[feature].value_counts() 

    # Count all transaction (both fraudulent and non fraudulent) for requested feature
    all_trans = dataset[dataset[feature].isin(fraud_cols)]
    count_all_trans = all_trans[feature].value_counts()

    # ratio of fraudulent transaction against total transaction by requsted feature
    most_fraudulent = (count_frauds / count_all_trans).sort_values(ascending=False)

    # create table 
    table = pd.concat([count_frauds, count_all_trans, most_fraudulent], axis='columns')
    table.columns = ['Fraudulent Transactions', 'Total Transaction', 'Ratio of Fraudulent to Total']
    return table.sort_values(by='Ratio of Fraudulent to Total', ascending=False)
```

### 3.5.1 Which Merchant has processed the most fraudulent transactions

```python
# get table of merchants who have processed fraudulent transactions
fraud_merchant = ask_fraudulent('merchant')
fraud_merchant.head()
```

```python
# plot the visual
fig, ax = plt.subplots(figsize = (15, 8))
sns.barplot(x=fraud_merchant['Ratio of Fraudulent to Total'], 
            y=fraud_merchant.index, 
            ax=ax, palette="Blues_d")
plt.title('Merchants with Most Fraudulent Transactions', fontsize=20)
plt.ylabel("Merchant")
plt.show()
```

Some merchants have as much as 90% and above of the total transactions they process being fraudulent. This is quite suspicious and it will be safe to blacklist such merchants.

### 3.5.2 Which customer account has been used to process the most Fraudulent Transactions

```python
# get table of customers whose accounts have processed fraudulent transactions
fraud_cust = ask_fraudulent('customer')
fraud_cust.head()
```

```python
# plot the visual of the top 20 most fraudulent customers
fig, ax = plt.subplots(figsize = (15, 8))
sns.barplot(x=fraud_cust.head(20)['Ratio of Fraudulent to Total'],
            y=fraud_cust.head(20).index, 
            ax=ax, palette="Blues_d")
plt.title('Top 20 Customers Account with the Most fraudulent Transactions', fontsize=20)
plt.ylabel("Customer")
plt.show()
```

Similar to Merchants, some accounts have 90% of its total transactions being fraudulent. There is a need to consider how accounts which are used for fraudulent transaction relate with merchants processing fraudulent transancations.

### 3.5.3 How do Customers Account with high frauds Relate with merchants

```python
fraud_cus = ask_fraudulent('customer').head(10).index
all_fraud = fraudulent_df[fraudulent_df['customer'].isin(fraud_cus)]

count_trans = all_fraud['customer'].value_counts()
count_merch =  all_fraud.groupby('customer')['merchant'].nunique()
freq_merch = all_fraud.groupby('customer').agg(lambda x:x.value_counts().index[0])['merchant']
table = pd.concat([count_trans, count_merch, freq_merch], axis='columns')
table.columns = ['Total Fraudulent Transactions', 'Different Merchants Used', 'Most Used Merchant']
table
```

Fraudsters generally switch merchants. Fraudsters use different merchants on customers accounts. It makes it difficult for them to be suspected. It would have been easier should all fraudulent transactions on an account be traceable to one or two merchants.
